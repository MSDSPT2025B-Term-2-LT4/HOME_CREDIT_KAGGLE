Features:
* paying_active_contract: Has active contract (0 or 1) -- if pmts_month_706T is NULL then 0 else 1 
* paying_closed_contract: Has closed contract (0 or 1) -- if pmts_month_158T is NULL then 0 else 1
* has_active_collateral: Active contract has collateral -- if pmts_month_706T is NOT NULL and collaterals_typeofguarante_669M is NOT NULL then 1 else 0
* has_closed_collateral: Closed contract has collateral -- if pmts_month_159T is NULL and collaterals_typeofguarante_359M is NOT NULL THEN 1 else 0
* valtype_active_collat: collater_typofvalofguarant_298M - Collateral Valuation type of active contract (optional)
* type_active_collat: collaterals_typeofguarante_669M - Collateral type for the active contract (optional)
* total_existing_contracts: Based on max numgroup1
* total_num_payments_made: countall rows
* avg_dpd_active: avg of pmts_dpd_1073P
* avg_dpd_terminated: avg of pmts_dpd_303P
